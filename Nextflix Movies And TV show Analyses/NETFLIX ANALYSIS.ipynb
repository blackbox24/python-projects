```python
import pandas as pd
```

```python
def wrangle(filename:str) -> pd.DataFrame:
    df = pd.read_csv(filename)
    # drop unnecessary columns
    columns_to_drop = []
    columns = ["type","date_added","release_year","rating", "duration", "country"]

    df = df[columns]

    # remove NaN's
    df.dropna(how="any", inplace=True)

    # split the duration
    df[["duration_value","duration_type"]] = df["duration"].str.split(" ", expand=True)
    df["duration_value"] = pd.to_numeric(df["duration_value"], errors="coerce")
    df["duration_type"] = df["duration_type"].replace(
        {
            "min":"Minutes",
            "season
    columns_to_drop.append("duration")

    # Split and strip whitespace from country
    df['country'] = df['country'].str.split(',').str[0].str.strip()

    # replace values of ratings
    df['content_rating'] = df['rating'].replace({
        'TV-MA': 'Mature Adults',
        'TV-14': '14+ Teens',
        'R': 'Restricted',
        'TV-PG': 'Parental Guidance',
        'PG-13': 'Parents Strongly Cautioned',
        'PG': 'Parental Guidance',
        'TV-G': 'General Audience',
        'TV-Y7': 'Kids 7+',
        'TV-Y': 'Children',
        'NR': 'Not Rated',
        'G': 'General',
        'TV-Y7-FV': 'Kids Fantasy Violence',
        'UR': 'Unrated',
        'NC-17': 'Adults Only'
    })
    # drop ratings column
    columns_to_drop.append("rating")
    df.drop(columns=columns_to_drop, inplace=True)

    # Parse Date
    df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")

    # Drop duplicates
    df.drop_duplicates(inplace=True)
    return df
    
```

```python
data = wrangle("./netflix_titles.csv")
```
