```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("./data/salesdaily.csv")
```

```python
data.head()
```

## QUESTIONS

- What are the total sales quantities for each drug category (ATC code)?

- Which individual drug brands have the highest total sales?

- Which three drugs have the highest sales in January 2015, July 2016, September 2017.

- Which drug has sold the most often in 2017?

- Which drug category has the highest average daily sales?

- Are respiratory drugs (R03) sold more during specific months?

```python
# What are the total sales quantities for each drug category (ATC code)?
ATC_CODE = ['M01AB', 'M01AE', 'N02BA', 'N02BE', 'N05B', 'N05C', 'R03','R06']
total_sales = data[ATC_CODE].sum()
```

```python
total_sales.sort_values(ascending=True)
```

```python
# Which individual drug brands have the highest total sales?
total_sales.max()
```

```python
#  Which three drugs have the highest sales in January 2015, July 2016, September 2017.
data["datum"] = pd.to_datetime(data["datum"])
```

```python
data.head()
```

```python
df = data.copy()
```

```python
conditions = (
    (df['datum'].dt.year == 2015) & (df['datum'].dt.month == 1) |  # Jan 2015
    (df['datum'].dt.year == 2016) & (df['datum'].dt.month == 7) |  # Jul 2016
    (df['datum'].dt.year == 2017) & (df['datum'].dt.month == 9)    # Sep 2017
)
filter_dates = df[conditions]
```

```python
filter_dates[ATC_CODE].sum().sort_values(ascending=False)[:3]
```

```python
# Which drug has sold the most often in 2017?
most_sold = df[df['datum'].dt.year == 2017]
```

```python
most_sold[ATC_CODE].sum().sort_values()[:1]
```
